```python
b2=10*a
plt.figure(1)
plt.plot(a,b,'r--')
plt.figure(2)
plt.plot(a,b2,'b--')
```

```python
t = np.arange(0,12,0.5)
y1=np.sin(t)
y2=np.cos(t)
plt.plot(t,y1,'y--',marker = 'p')
plt.plot(t,y2,'r--',marker='*')
plt.xlabel("Time")
plt.ylabel("Amp")
plt.title("SinCos")
plt.legend(["Sin","Cos"])
```

```python
x=np.arange(0,5,1)
y1=x+1
y2=x+2
y3=x+3
y4=x+4
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,'r--',marker='*')
plt.plot(x,y2,'g',marker='o')
plt.xlabel("Time")
plt.ylabel("Amp")
plt.title("SinCos")
plt.legend(["Sin","Cos"])
plt.subplot(1,2,2)
plt.plot(x,y3,'b--',marker='x')
plt.plot(x,y4,'y',marker='v')
plt.xlabel("Time")
plt.ylabel("Amp")
plt.title("SinCos")
plt.legend(["Sin","Cos"])
```
